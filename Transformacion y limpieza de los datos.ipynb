```python
#VAMOS A VISUALIZAR NUESTRO DATASET
df_viviendas_california
```

```python
#DEBIDO A QUE TENEMOS DATOS MUY GRANDES Y ALEJADOSDE LA MEDIA, PARA EVITAR UN ALTO SESGO PROCEDEROS A REEMPLAZAR 
#LOS VALORES NULOS CON LA MEDIANA
df_viviendas_california.fillna(435, inplace=True)
```

```python
#A CONTINUACION PROCEDEREMOS A CAMBIAR EL NOMBRE DE LAS COLUMNAS
df_viviendas_california.rename(columns={'longitude': 'Longitud', 
                                'latitude': 'Latitud', 
                                'housing_median_age': 'Edad media de vivienda',
                                'total_rooms': 'Habitaciones en total',
                                'total_bedrooms': 'Dormitorios en total',
                                'population': 'Poblacion',
                                'households': 'Cantidad de hogares',
                                'median_income': 'Ingreso medio',
                                'median_house_value': 'Valor medio de viviendas',
                                'ocean_proximity': 'Proximidad al oceano'}, inplace=True)
```

```python
df_viviendas_california.head(50)
```

```python
#AHORA PROCEDEMOS A CAMBIAR EL TIPO DE DATO DE LAS COLUMNAS
df_viviendas_california['Dormitorios en total'] = df_viviendas_california['Dormitorios en total'].astype('Int64')
df_viviendas_california['Edad media de vivienda'] = df_viviendas_california['Edad media de vivienda'].astype('Int64')
df_viviendas_california['Habitaciones en total'] = df_viviendas_california['Habitaciones en total'].astype('Int64')
df_viviendas_california['Cantidad de hogares'] = df_viviendas_california['Cantidad de hogares'].astype('Int64')
```

```python
df_viviendas_california.info()
```

```python
#AQUI PROCEDEMOS A CAMBIAR LAS FILAS DE LA COLUMNA PROXIMIDAD AL OCEANO A MINUSCULAS.
df_viviendas_california['Proximidad al oceano'] = df_viviendas_california['Proximidad al oceano'].str.capitalize()
```
